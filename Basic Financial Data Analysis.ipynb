```python
goog["daily_return"]= (goog["close"]/goog["close"].shift(1))-1
```

```python
def cal_daily_return(*args):
    for dataset in args:
        dataset["daily_return"]= (dataset["close"]/dataset["close"].shift(1))-1
    
```

```python
cal_daily_return(goog,amzn,msft,aapl)
```

```python
def histogram_daily_return(labels, *args):
    coccat_list = []
    for dataset in args:
        concat_list.append(dataset["daily_return"])
        
    merged_df = pd.concat(concat_list, axix=1)
    merged_df.columns = labels
    for col in merged_df.columns:
        merged_df[col].hist(bins=50,label=col,figsize=(10,6))
    plt.legend()    
    return merged_df
```

```python
df=histogram_daily_return(labels =["Apple","Google","Amazon","Microsoft"], aapl, goog, amzn, msft)
```

```python
# using function
def plot_histogram(labels,*args):
    concat_list=[]
    for dataset in args:
        concat_list.append(dataset["daily_return"])
        
    merged_df=pd.concat(concat_list,axis=1)
    merged_df.columns=labels
#         dataset["daily_return"].hist(bins=50,label=labels[index])
    for col in merged_df.columns:
        merged_df[col].hist(bins=50,label=col,figsize=(10,5))
    plt.legend()
    return merged_df
```

```python
df =plot_histogram(["Apple","Google","Amazon","Microsoft"], aapl, goog, amzn, msft)
```

```python
# another way
def plot_histogram2(labels,*args):
    index=-1
    for dataset in args:
        index+=1
        dataset["daily_return"].hist(bins=50,label=labels[index])
    plt.legend()
```

```python
df=plot_histogram2(
    ["Apple","Google","Amazon","Microsoft"],
    aapl,goog,amzn,msft)
```
